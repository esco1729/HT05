### Ciudades con las menores medianas para cada variable por año

```python
grouped_means = df.groupby(['Ubicacion', 'Anio']).median().reset_index()
min_locations = {}
max_locations = {}

for variable in [col for col in grouped_means.columns if col not in ['Ubicacion', 'Anio']]:
    pivot_data = grouped_means.pivot(index='Anio', columns='Ubicacion', values=variable)

    min_locations[variable] = pivot_data.idxmin(axis=1)
    max_locations[variable] = pivot_data.idxmax(axis=1)

min_locations_df = pd.DataFrame(min_locations)
max_locations_df = pd.DataFrame(max_locations)

print("Ciudades con las menores medianas para cada variable por año")
print(min_locations_df.reset_index())
```

### Ciudades con las mayores medianas para cada variable por año

```python
print("\Ciudades con las mayores medianas para cada variable por año")
print(max_locations_df.reset_index())
```

```python
df.groupby(['Ubicacion', 'Anio']).sum().reset_index()
min_locations = {}
max_locations = {}

for variable in [col for col in df.groupby(['Ubicacion', 'Anio']).sum().reset_index().columns if col not in ['Ubicacion', 'Anio']]:
    pivot_data = df.groupby(['Ubicacion', 'Anio']).sum().reset_index().pivot(index='Anio', columns='Ubicacion', values=variable)

    min_locations[variable] = pivot_data.idxmin(axis=1)
    max_locations[variable] = pivot_data.idxmax(axis=1)
min_locations_df = pd.DataFrame(min_locations)
max_locations_df = pd.DataFrame(max_locations)
```

```python
df.drop('Fecha', axis=1 , inplace = True)
```

```python
label = {'No': 0, 'Yes': 1}
df["Lluvia mañana"].replace(label, inplace = True)
```

```python

X = df.drop('Lluvia mañana', axis = 1)
Y = df["Lluvia mañana"]

X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numericX = X.select_dtypes(include=[np.number])
d_scaled = scaler.fit_transform(numericX)
data_scaled1 = pd.DataFrame(d_scaled)
data_scaled1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(d_scaled, Y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C = 0.1, max_iter = 500)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
X_test
```

```python
y_pred
```

```python
predicted_proba = model.predict_proba(X_test)

positive_proba = predicted_proba[:, 1]
positive_proba
```

```python
print('Weight Coefficient : ', model.coef_)
print('Bias : ', model.intercept_)
```

```python
print('Test accuracy: ', model.score(X_test, y_test))
print('Train accuracy: ', model.score(X_train, y_train))
```
